```python
#The required Imports
import pandas as pd
import numpy as np 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the csv Data in Python
database=pd.read_csv("/Users/suvradattabanik/Downloads/Imports_Exports_Dataset.csv")
```

```python
database
```

```python
# Sampled 2001 rows randomly from the given data set
samData = database.sample(n=2001,random_state=55049)
```

```python
samData
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country_value = samData.groupby('Country')['Value'].sum().reset_index()
country_quantity = samData.groupby('Country')['Quantity'].sum().reset_index()

# Aggregate data by Product
product_value = samData.groupby('Product')['Value'].sum().reset_index()
product_quantity = samData.groupby('Product')['Quantity'].sum().reset_index()

# Get top 5 countries and products by value and quantity
top_countries_value = country_value.nlargest(5, 'Value')
top_countries_quantity = country_quantity.nlargest(5, 'Quantity')

top_products_value = product_value.nlargest(5, 'Value')
top_products_quantity = product_quantity.nlargest(5, 'Quantity')

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# Bar chart for top countries by value
plt.subplot(2, 2, 1)
sns.barplot(x='Value', y='Country', data=top_countries_value, palette='viridis')
plt.title('Top Countries by Total Value')
plt.xlabel('Total Value')
plt.ylabel('Country')

# Bar chart for top countries by quantity
plt.subplot(2, 2, 2)
sns.barplot(x='Quantity', y='Country', data=top_countries_quantity, palette='viridis')
plt.title('Top Countries by Total Quantity')
plt.xlabel('Total Quantity')
plt.ylabel('Country')

# Bar chart for top products by value
plt.subplot(2, 2, 3)
sns.barplot(x='Value', y='Product', data=top_products_value, palette='viridis')
plt.title('Top Products by Total Value')
plt.xlabel('Total Value')
plt.ylabel('Product')

# Bar chart for top products by quantity
plt.subplot(2, 2, 4)
sns.barplot(x='Quantity', y='Product', data=top_products_quantity, palette='viridis')
plt.title('Top Products by Total Quantity')
plt.xlabel('Total Quantity')
plt.ylabel('Product')

# Show the plots
plt.tight_layout()
plt.show()
```

What is the trend of trade values over time?
Plot: Line chart to visualize how the total trade values have changed over the months/years.

```python
# Convert 'Date' to datetime format
samData['Date'] = pd.to_datetime(samData['Date'], format='%d-%m-%Y')

# Group by month and year, summing the trade values
monthly_trend = samData.resample('M', on='Date').sum().reset_index()

# Set seaborn style
sns.set(style="whitegrid")

# Plotting the trend of trade values over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trend, x='Date', y='Value', marker='o')
plt.title('Trend of Trade Values Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Trade Value', fontsize=12)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()
plt.show()
```

What payment terms are most common for exports versus imports?
Plot: Stacked bar chart showing the distribution of payment terms for imports and exports.

```python
# Create a DataFrame to count payment terms for imports and exports
payment_terms_counts = samData.groupby(['Payment_Terms', 'Import_Export']).size().unstack(fill_value=0)

# Set seaborn style
sns.set(style="whitegrid")

# Plotting the Stacked Bar Chart
plt.figure(figsize=(10, 6))
payment_terms_counts.plot(kind='bar', stacked=True, color=sns.color_palette("Set2"))
plt.title('Distribution of Payment Terms for Imports and Exports', fontsize=14)
plt.xlabel('Payment Terms', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()
plt.legend(title='Import/Export')
plt.show()
```

How does the weight of goods correlate with their trade value?
Plot: Scatter plot showing the correlation between weight and trade value, colored by import/export status.

```python
# Set seaborn style
sns.set(style="whitegrid")

# Plotting the Scatter Plot for Weight vs Trade Value
plt.figure(figsize=(10, 6))
sns.scatterplot(data=samData, x='Weight', y='Value', hue='Import_Export', palette='Set1', s=100)
plt.title('Correlation between Weight and Trade Value', fontsize=14)
plt.xlabel('Weight of Goods', fontsize=12)
plt.ylabel('Trade Value', fontsize=12)
plt.axhline(0, color='gray', linestyle='--')  # Reference line at y=0 (optional)
plt.tight_layout()
plt.show()
```

```python
# Count the number of transactions for each customer
customer_counts = samData['Customer'].value_counts()

# Filter for repeating customers (those with more than one transaction)
repeating_customers = customer_counts[customer_counts > 1]

# Set seaborn style
sns.set(style="whitegrid")

# Plotting the count of repeating customers
plt.figure(figsize=(10, 6))
sns.barplot(x=repeating_customers.index, y=repeating_customers.values, palette='Set2')
plt.title('Count of Transactions for Repeating Customers', fontsize=14)
plt.xlabel('Customer', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()
plt.show()
```

Which countries are the top 3 importers versus exporters?

```python
# Convert 'Date' to datetime format
samData['Date'] = pd.to_datetime(samData['Date'])

# Extract year and month for aggregation
samData['Year'] = samData['Date'].dt.year
samData['Month'] = samData['Date'].dt.month

# Group by Country and Import/Export status to sum trade values
monthly_trade = samData.groupby(['Country', 'Import_Export'])['Value'].sum().reset_index()

# Find the top 3 importers and exporters
top_exporters = monthly_trade[monthly_trade['Import_Export'] == 'Export'].nlargest(3, 'Value')
top_importers = monthly_trade[monthly_trade['Import_Export'] == 'Import'].nlargest(3, 'Value')

# Combine the top exporters and importers
top_countries = pd.concat([top_exporters, top_importers])

# Set seaborn style
sns.set(style="whitegrid")

# Plotting the grouped bar chart for top importers and exporters
plt.figure(figsize=(10, 6))
sns.barplot(data=top_countries, x='Country', y='Value', hue='Import_Export', palette='Set2')
plt.title('Top 3 Importers and Exporters', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Trade Value', fontsize=12)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()
plt.show()
```

Which countries are the least 3 importers versus exporters?

```python
# Convert 'Date' to datetime format
samData['Date'] = pd.to_datetime(samData['Date'])

# Group by Country and Import/Export status to sum trade values
monthly_trade = samData.groupby(['Country', 'Import_Export'])['Value'].sum().reset_index()

# Find the least 3 importers and exporters
least_exporters = monthly_trade[monthly_trade['Import_Export'] == 'Export'].nsmallest(3, 'Value')
least_importers = monthly_trade[monthly_trade['Import_Export'] == 'Import'].nsmallest(3, 'Value')

# Combine the least exporters and importers
least_countries = pd.concat([least_exporters, least_importers])

# Set seaborn style
sns.set(style="whitegrid")

# Plotting the grouped bar chart for least importers and exporters
plt.figure(figsize=(10, 6))
sns.barplot(data=least_countries, x='Country', y='Value', hue='Import_Export', palette='Set2')
plt.title('Least 3 Importers and Exporters', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Trade Value', fontsize=12)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()
plt.show()
```

What is the distribution of trade values across different categories?
Plot: Dist plot to visualize the distribution of trade values for a specific category (or overall) with a kernel density estimate (KDE).

```python
# Set seaborn style
sns.set(style="whitegrid")

# Plotting the distribution of trade values for a specific category (e.g., 'Clothing')
plt.figure(figsize=(12, 6))
sns.histplot(data=samData[samData['Category'] == 'Clothing'], x='Value', kde=True, bins=10, color='blue')
plt.title('Distribution of Trade Values for Clothing', fontsize=14)
plt.xlabel('Trade Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Set seaborn style
sns.set(style="whitegrid")

# Filter the DataFrame for electronics and furniture goods
filtered_data = samData[samData['Category'].isin(['Electronics', 'Furniture'])]

# Joint Plot for Trade Value vs Weight for Electronics and Furniture
sns.jointplot(data=filtered_data, x='Weight', y='Value', kind='scatter', hue='Category', palette='Set1')
plt.suptitle('Relationship between Trade Value and Weight of Electronics and Furniture Goods', y=1.02)  # Title with offset
plt.xlabel('Weight of Goods', fontsize=12)
plt.ylabel('Trade Value', fontsize=12)
plt.show()
```
